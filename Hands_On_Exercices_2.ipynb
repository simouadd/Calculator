```python
# 1
import numpy as np
import matplotlib.pyplot as plt


def draw_circle(x,y,r):
    t= np.linspace(0,7,100)
    X=x+(r*np.cos(t))
    Y=y+(r*np.sin(t))
    plt.plot(X,Y)
    plt.axis("equal")

draw_circle(20,20,12)
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt


def draw_centric_circle(x,y,r,n):
    for i in range(0,n):
        draw_circle(x,y,r)
        r=r-2

draw_centric_circle(10,10,10,5)
```

```python
# 3
import numpy as np
import matplotlib.pyplot as plt


def draw_rectangle_circle(x,y,r,h,w):
    y1=y

    for i in range(0,h):
        for i in range(0,w):
            draw_circle(x,y,r)
            y=y-2*r
        y=y1
        x=x-2*r

draw_rectangle_circle(0,10,2,7,5)
```

```python
# 4
def draw_circeles_x(r,n):
    x=0
    y=0

    for i in range(0,n):
        draw_circle(x,y,r)
        r0=r
        r=r-r/2
        R=r0+r    # R is the sum of the r of the last circle and the r of the next one
        x=x+R


def draw_circeles_y(r,n):
    x=0
    y=0

    for i in range(0,n):
        draw_circle(x,y,r)
        r0=r
        r=r-r/2
        R=r0+r    # R is the sum of the r of the last circle and the r of the next one
        y=y-R


draw_circeles_x(10,5)
draw_circeles_y(10,5)
```

```python
# 5
import numpy as np
import matplotlib.pyplot as plt
def draw_triangle_circles(r,w):
    x=0
    y=0
    x0=0
    while w!=0:

        for i in range(0,w):
            draw_circle(x,y,r)
            x=x+2*r
        y=y+np.sqrt(3)*r
        x=x0+r
        x0=x
        w=w-1
draw_triangle_circles(5,9)
```

```python
# 6

def draw_circles_k(x,y,r,n):

    draw_circle(x,y,r)

    if n>0:
        x1=x+r+r/2
        y1=y-r-r/2

        r=r/2
        n=n-1
        draw_circles_k(x,y1,r,n)
        draw_circles_k(x1,y,r,n)



draw_circles_k(0,0,6,4 )
```

```python
# 7

def draw_circles4(x,y,r,n,direction):
    draw_circle(x,y,r)
    if n>0:
        x1=x+r+r/2
        y1=y-r-r/2
        x2=x-r-r/2
        y2=y+r+r/2
        r=r/2
        n=n-1
        for d in direction:
            if d==0:
                draw_circles4(x,y2,r,n,[0,1,3])
            elif d==1:
                draw_circles4(x1,y,r,n,[0,1,2])

            elif d==2:
                draw_circles4(x,y1,r,n,[1,2,3])
            elif d==3:
                draw_circles4(x2,y,r,n,[0,2,3])



draw_circles4(0,0,3,5,[0,1,2,3])
```
